```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('SampleSuperstore.csv')
```

```python
# Display basic information about the dataset
print("Dataset shape:", data.shape)
print("\nColumn names:", data.columns)
print("\nData types:\n", data.dtypes)
print("\nFirst few rows:\n", data.head())
```

```python
# Summary statistics
print("\nSummary statistics:\n", data.describe())
```

```python
print("Dataset Structure and Dimensions:")
print(data.info())
```

```python
# Univariate Analysis
# Ship Mode Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Ship Mode', data=data)
plt.title('Ship Mode Distribution')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```

The "Standard class" is the most used mode.

```python
# Segment Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Segment', data=data)
plt.title('Segment Distribution')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.show()
```

The customer segmenr "Consumer" is the most frequent.



```python
# Category Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=data)
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

The most common product category is "Office Supplies Category".

```python
# Sub-Category Distribution
plt.figure(figsize=(14, 8))
sns.countplot(y='Sub-Category', data=data, order=data['Sub-Category'].value_counts().index)
plt.title('Sub-Category Distribution')
plt.xlabel('Count')
plt.ylabel('Sub-Category')
plt.show()
```

"Binders" is the most common sub-category.
"Copiers" is the least common sub-category.

```python
# Bivariate Analysis
# Sales vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=data)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

This scatter plot shows the relationship between sales and profit. It helps to identify trends and outliers, such as high sales with low profit or vice versa.

```python
# Sales by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Segment', y='Sales', data=data)
plt.title('Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.show()
```

This box plot shows the distribution of sales across different customer segments. It helps to compare sales performance between segments.


```python
# Profit by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Segment', y='Profit', data=data)
plt.title('Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()
```

This box plot shows the distribution of profit across different customer segments. It helps to compare profitability between segments.

```python
# Sales by Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Sales', data=data)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

This box plot shows the distribution of sales across different product categories. It helps to compare sales performance between categories.

```python
# Profit by Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Profit', data=data)
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

This box plot shows the distribution of profit across different product categories. It helps to compare profitability between categories.

```python
# Correlation Analysis
# Selecting only numeric columns
import numpy as np
numeric_data = data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

This code selects only numeric columns from the dataset and generates a heatmap to show the correlation matrix. It helps to identify the strength and direction of relationships between numerical variables.

```python
# Multivariate Analysis
# Profit vs Discount vs Segment
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Discount', y='Profit', hue='Segment', data=data)
plt.title('Profit vs Discount by Segment')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.legend()
plt.show()
```

This scatter plot shows the relationship between discount and profit, with data points colored by customer segment. It helps to understand how discounts affect profitability in different segments.

```python
# Identify Weak Areas
# Find sub-categories with negative profit
negative_profit_sub_categories = data[data['Profit'] < 0]['Sub-Category'].unique()
print("\nSub-Categories with Negative Profit:")
print(negative_profit_sub_categories)
```

This code identifies product sub-categories that have negative profit, which indicates potential problem areas where the company is losing money.

```python
# Find regions with low average profit
average_profit_by_region = data.groupby('Region')['Profit'].mean().sort_values()
print("\nAverage Profit by Region:")
print(average_profit_by_region)
```

This code calculates the average profit by region and sorts the results. It helps to identify regions with low profitability, which may require further investigation.
